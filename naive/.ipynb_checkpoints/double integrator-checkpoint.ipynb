# Double Integrator System

See https://en.wikipedia.org/wiki/Double_integrator . It models the dynamics of a simple mass in one-dimensional space under the effect of a time-varying force input $u$.

Original continuous model

$\dot{x}=y,\dot{y}=p$, therefore $A\doteq\begin{pmatrix}0&1\\0&0\end{pmatrix}$ and $B\doteq\begin{pmatrix}0\\1\end{pmatrix}$.

The continuous system will be discretized as
$$(\forall n\in\mathbb{N})\quad x_{n+1} = x_n + \gamma y_n + \frac{1}{2}\gamma^2p, y_{n+1} = y_n + \gamma p,$$
where $\gamma\in\mathbb{R}_{++}$ and  $p\in P\doteq[-\mu,+\mu]$ with $\mu\in\mathbb{R}_{++}$. Thus,

$(\forall n\in\mathbb{N})\quad C_{n+1}\doteq (\text{Id}-\gamma A)(C_n + P)\cap C_0,$

where $C_0\doteq [-a,+a]\times[-a,+a]$, with $a\in\mathbb{R}_{++}$

### Discretisation with Scipy

```python
from scipy import signal as sig
import numpy as np
```

```python
gamma = 1./10.
```

```python
sys = sig.StateSpace(np.array([[0, 1], [0, 0]]), np.array([[0], [1]]), np.array([[1, 0]]), np.array([[0]]))
sys = sys.to_discrete(gamma)
```

```python
A = matrix(np.ascontiguousarray(sys.A))
print(A)
Ainv = A.inverse()
print(Ainv)
```

```python
B = matrix(np.ascontiguousarray(sys.B))
B
```

### Discretisation  Laengle

Let $\varphi$ a set-valued operator and $\gamma\in\mathbb{R}_{++}$. The **resolvent** of $\varphi$ is
$$J_{\varphi}\doteq(\text{Id}+\varphi)^{-1}$$
and the **Yosida approximation** of $\varphi$ of index $\gamma$ is
$$^{\gamma}\varphi\doteq\frac{1}{\gamma}(\text{Id}-J_{\gamma\varphi}).$$

```python
A = matrix([[0, 1], [0, 0]])
J = (matrix.identity(2) + gamma*A).inverse()   #Resolvent
yosA = 1/gamma*(matrix.identity(2) - J)
B = matrix([[0, 0], [0, 1]])
print('A = ', A)
print('J = ', J)
print('yosA = ', yosA)
print('B = ', B)
```

```python
a = 1.0
C = Polyhedron(vertices = [[+a, -a], [+a, +a], [-a, +a], [-a, -a]])
```

```python
mu = 0.1           # 0.1, 0.3, 1.0, 2.0
P = Polyhedron(vertices = [[0, -mu], [0, +mu]])
gammaP = gamma*P
gammaP
```

```python
def exact(C0, gammaP = Polyhedron, N = int, J = matrix):
    """
    Calculate the "exact" viable polyhedra of a system dx/dt = Ax + Bp
    
    C0, P: Sage Polydedra of same dimension, P is the discretised approximation
    N    : Positive integer
    Ainv : Sage matrix of the same dimension (discretised approximation)
    """
    
    C, n = C0, 0
    while n <= N:
        print(n, 'volume = ', C.volume(), end = '\r')
        # print(n, 'volume = ', C.volume(), end = ' ')
        if C.is_empty():
            break
        Cp = (C.minkowski_sum(gammaP)).linear_transformation(J).intersection(C0)
        if Cp == C:
            break
        else:
            C = Cp
            n += 1
    print('')
    return(C)
```
