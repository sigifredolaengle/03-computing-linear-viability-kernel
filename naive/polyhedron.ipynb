```python
p = polytopes.cube()
A = Matrix([[0.8,0,0],[0,0.8,0],[0,0,0.8]]); A
(A*p).plot()
```

```python
(A*p).Hrepresentation(1).is_H()
```

```python
for i in range(2000):
    p = A*p
```

```python
print(p)
```

```python
p.to_linear_program()
```

```python
lp, x = p.to_linear_program(return_variable = True)
```

```python
lp.set_objective(-2*x[0] + 1*x[1] + 39*x[2])
lp.solve()
```

```python
lp.get_values(x)
```
